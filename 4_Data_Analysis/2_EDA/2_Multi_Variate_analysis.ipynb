```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../Data/top100_ai_tools_2026.csv")
df
```

```python
df.columns
```

# 1. Num Vs Num

## 1.1. Sketter plot

```python
sns.relplot(kind='scatter', data=df, x='Starting_Price_USD', y='User_Rating')
```

### 3D Scatter plot

```python
fig = plt.figure(figsize=(15,7))

ax = plt.subplot(projection="3d")
ax.scatter3D(df['Starting_Price_USD'], df['User_Rating'], df['Agentic_Capability_Score'])
ax.set_xlabel('Starting_Price_USD')
ax.set_ylabel('User_Rating')
ax.set_zlabel('Agentic_Capability_Score')
```

## 1.2. Lineplot

```python
sns.relplot(kind='line', data=df, x='Release_Year', y='Agentic_Capability_Score')
```

```python
sns.relplot(kind='line', data=df, x='Release_Year', y='Agentic_Capability_Score',hue='API_Available')
```
