## 4.3 Pairplot (only for numerical col)

```python
num = ['Starting_Price_USD', 'User_Rating', 'Agentic_Capability_Score', 'Active_Users_Est']

cat = ['Primary_Category','Pricing_Model', 'API_Available', 'Release_Year']
```

```python
df_num = df[num]
df_num['API_Available'] = df['API_Available']
df_num
```

```python
g = sns.pairplot(df_num, hue='API_Available')
g.map(sns.scatterplot)
```

```python
# diagonal -> histogram ; non-diagonal -> (lower -> scatterplot, upper->violin)
g = sns.pairplot(df_num, hue='API_Available')
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.violinplot)
```
