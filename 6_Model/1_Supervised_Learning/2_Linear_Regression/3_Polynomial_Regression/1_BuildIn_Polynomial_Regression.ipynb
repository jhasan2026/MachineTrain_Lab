```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = 8 * np.random.rand(200) - 4
y = (0.8 * X**2) + (0.7 * X) + 2 + np.random.randn(200)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train, X_test = pd.DataFrame(X_train), pd.DataFrame(X_test)
```

```python
plt.plot(X_train, y_train,"b.")
```

# Using Linear Regression

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train,y_train,"b.")
plt.plot(X_test,y_pred,"r-")
```

# Using Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=True)

X_train_trf = poly.fit_transform(X_train)
X_test_trf = poly.transform(X_test)
```

```python
lr2 = LinearRegression()
lr2.fit(X_train_trf, y_train)
y_pred_trf = lr2.predict(X_test_trf)
r2_score(y_test,y_pred_trf)
```

```python
X_graph = np.linspace(-4,4,200).reshape(200,1)
y_pred_graph = lr2.predict(poly.transform(X_graph))
```

```python
plt.plot(X_train,y_train,"b.")
plt.plot(X_graph,y_pred_graph,"r-")
```
