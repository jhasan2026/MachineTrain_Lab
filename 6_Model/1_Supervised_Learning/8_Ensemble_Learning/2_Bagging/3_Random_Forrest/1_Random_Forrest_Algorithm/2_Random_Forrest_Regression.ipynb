```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate training data
X_train = np.sort(np.random.rand(150) * 10 - 5)
y_train = np.exp(-X_train ** 2) + 1.5 * np.exp(-(X_train - 2) ** 2) + np.random.normal(0.0, 0.1, 150)

# Generate testing data
X_test = np.sort(np.random.rand(100) * 10 - 5)
y_test = np.exp(-X_test ** 2) + 1.5 * np.exp(-(X_test - 2) ** 2) + np.random.normal(0.0, 0.1, 100)

# Define the true function
f = lambda x: np.exp(-x ** 2) + 1.5 * np.exp(-(x - 2) ** 2)
```

```python
X = pd.concat([pd.DataFrame(X_train,columns=['Feature']), pd.DataFrame(X_test,columns=['Feature'])], axis=0)
y = pd.concat([pd.DataFrame(y_train,columns=['Value']), pd.DataFrame(y_test,columns=['Value'])], axis=0)
X
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X_train, y_train, s=5)
plt.plot(X_test,f(X_test), c='red')
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
from sklearn.base import clone

def evaluateModelR2(model):
    copyModel = clone(model)
    model.fit(X_train.reshape(-1,1), y_train)
    y_pred = model.predict(X_test.reshape(-1,1))
    return {
        "R2": r2_score(y_test, y_pred),
        "Cross Val Score": (cross_val_score(copyModel, X,y,scoring='r2',cv=5)).mean()
    }
```

```python
from sklearn.ensemble import RandomForestRegressor


rf = RandomForestRegressor(
    n_estimators=20,
    max_features=0.6,
    max_depth=8,
    max_samples=0.75
)
print(evaluateModelR2(rf))
```

# All possible case : Grid Search CV

```python
params = {
    'n_estimators' : [20,60],
    'max_features' : [0.4,0.6,1.0],
    'max_depth' : [2,8,None],
    'max_samples' : [0.5,0.75,1.0]
}
```

```python
from sklearn.model_selection import ParameterGrid

grid = ParameterGrid(params)
print(len(grid))
```

```python
from sklearn.model_selection import GridSearchCV

rf_grid = GridSearchCV(
    estimator=RandomForestRegressor(),
    param_grid= params,
    cv=5,
    verbose=2,
    n_jobs=-1
)
```

```python
rf_grid.fit(X_train.reshape(-1,1),y_train.reshape(-1,1))
```

```python
rf_grid.best_score_
```

```python
rf_grid.best_params_
```

```python
best_model = rf_grid.best_estimator_
print(evaluateModelR2(best_model))
```

# Plotting

```python
def plotRegion(classfier_model):
    plt.figure(figsize=(12, 8))
    plt.scatter(X_train, y_train, s=5)
    plt.plot(X_test,f(X_test), c='red', label="Actual Line")
    
    y_pred = classfier_model.predict(X_train.reshape(-1,1))
    plt.plot(X_train, y_pred, "g", lw=2, label="RF prediction")
    plt.legend()
    plt.show()
```

```python
plotRegion(best_model)
```
