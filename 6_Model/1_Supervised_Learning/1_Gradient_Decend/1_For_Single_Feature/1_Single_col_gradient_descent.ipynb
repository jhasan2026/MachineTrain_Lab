```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
X
```

```python
plt.scatter(X,y)
```

# 1. Check Linear Reg alogo s regression line

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)

intercept = lr.intercept_
coef = (lr.coef_)[0]
```

```python
y_pred = (X * lr.coef_) + intercept
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred, color='r')
```

# Using Gradient Changing

```python
import random 
def random_color_generate():
    n = random.randint(0,0xFFFFFF)
    return f"#{n:06x}"
```

```python
def gradient_change(epochs, lr):
    coef = 78.35
    intercept = 100
    plt.scatter(X, y)
    for epoch in range(epochs):
        dL_by_bd = -2 * np.sum(y - coef*(X.ravel()) - intercept)
        intercept = intercept - lr*dL_by_bd
        y_pred = (coef * X) + intercept
        plt.plot(X,y_pred, color='{}'.format(random_color_generate()),label=f"Epoch: {epoch}")
        
    plt.legend()
```

```python
gradient_change(10,0.01)
```
