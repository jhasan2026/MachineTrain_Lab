### e) 3D Surface Plots (for plane)

```python
# x^2 + y^2
xx , yy = np.meshgrid(np.linspace(-10,10,100), np.linspace(-10,10,100))

z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection="3d")

surface_obj = ax.plot_surface(xx,yy, z, cmap="viridis")
fig.colorbar(surface_obj)
```

```python
# sinx
xx , yy = np.meshgrid(np.linspace(-10,10,100), np.linspace(-10,10,100))

z = np.sin(xx)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection="3d")

surface_obj = ax.plot_surface(xx,yy, z, cmap="viridis")
fig.colorbar(surface_obj)
```

```python
# z = sinx + cosy
xx , yy = np.meshgrid(np.linspace(-10,10,100), np.linspace(-10,10,100))

z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection="3d")

surface_obj = ax.plot_surface(xx,yy, z, cmap="viridis")
fig.colorbar(surface_obj)
```

# 4. Subplot

```python
fig, ax = plt.subplots(figsize=(15,6))

ax.scatter(football['SoT'],football['Goals'])
ax.set_title("Shot on Target VS Goals")
ax.set_xlabel("Short on Target")
ax.set_ylabel("Goals")
fig.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,sharex=True,  figsize=(15,10))


ax[0].scatter(football['SoT'],football['Goals'])
ax[0].set_title("Shot on Target VS Goals")
ax[0].set_ylabel("Goals")

ax[1].scatter(football['SoT'],football['Shots'])
ax[1].set_title("Shot on Target VS Shots")
ax[1].set_ylabel("Shots")

ax[1].set_xlabel("Short on Target")

fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,  figsize=(15,10))


ax[0].scatter(football['SoT'],football['Goals'])
ax[0].set_title("Shot on Target VS Goals")
ax[0].set_xlabel("Short on Target")
ax[0].set_ylabel("Goals")

ax[1].scatter(football['SoT'],football['Shots'])
ax[1].set_title("Shot on Target VS Shots")
ax[1].set_xlabel("Short on Target")
ax[1].set_ylabel("Shots")


fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,sharex=True,  figsize=(15,12))


ax[0,0].scatter(football['SoT'],football['Goals'])
ax[0,0].set_title("Shot on Target VS Goals")
ax[0,0].set_xlabel("Short on Target")
ax[0,0].set_ylabel("Goals")

ax[0,1].scatter(football['SoT'],football['Shots'])
ax[0,1].set_title("Shot on Target VS Shots")
ax[0,1].set_xlabel("Short on Target")
ax[0,1].set_ylabel("Shots")

ax[1,0].hist(football['SoT'])
ax[1,0].set_xlabel("Short on Target")

ax[1,1].hist(football['Goals'])
ax[1,1].set_xlabel("Goals")

fig.show()
```

### one by one (alternative way)

```python
fig = plt.figure(figsize=(16,12))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(football['SoT'],football['Goals'])
ax1.set_title("Shot on Target VS Goals")
ax1.set_xlabel("Short on Target")
ax1.set_ylabel("Goals")

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(football['SoT'],football['Shots'])
ax2.set_title("Shot on Target VS Shots")
ax2.set_xlabel("Short on Target")
ax2.set_ylabel("Shots")

ax3 = fig.add_subplot(2,2,3)
ax3.hist(football['SoT'])
ax3.set_xlabel("Short on Target")
```
