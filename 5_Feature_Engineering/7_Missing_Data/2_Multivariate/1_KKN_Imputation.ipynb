# KNN Imputation (Only numerical Data)

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.preprocessing import OrdinalEncoder

cat_pipe = Pipeline([
    ("cat_imp", SimpleImputer(strategy='most_frequent')),
    ("encod",OrdinalEncoder(categories=[['Po','TA','Fa','Gd','Ex']]))
])

imp_trf = ColumnTransformer(transformers=[
    ("num_imp", KNNImputer(n_neighbors=3, weights='distance'), ['LotFrontage']),
    ("cat_pipe", cat_pipe, ['GarageQual'])
])
```

```python
X_train_trf = imp_trf.fit_transform(X_train)
X_test_trf = imp_trf.transform(X_test)
```
