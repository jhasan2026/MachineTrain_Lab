```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../Data/cars.csv")
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['selling_price']), df['selling_price'], test_size=0.2, random_state=42)
```

```python
X_train_encod = X_train[['fuel','owner']]
X_test_encod = X_test[['fuel','owner']]
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse_output=False, dtype=np.int32)

X_train_encod = ohe.fit_transform(X_train_encod)
X_test_encod = ohe.transform(X_test_encod)
```
