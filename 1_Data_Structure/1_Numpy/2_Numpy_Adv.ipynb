# Boolean Indexing

```python
arr > 10
```

```python
arr[ arr > 10]
```

```python
arr2 = np.random.randint(1,25 , (4,6))
arr2
```

```python
arr2 % 2 == 0
```

```python
(arr2 % 2 == 0) & (arr2 > 10)
```

```python
arr2[ arr2 % 2 == 0 ]
```

```python
arr2[ (arr2 % 2 == 0) & (arr2 > 10) ]
```

# Missing Value Check

```python
misArr = np.array([
    [1,np.nan],
    [3,4]
])
misArr
```

```python
np.isnan(misArr)
```

```python
misArr[ np.isnan(misArr) ]
```

```python
misArr[ ~ np.isnan(misArr) ]
```

# Broadcasting

```python
arr1 = np.random.randint(1,20, (3,4))
arr2 = np.random.randint(1,20, (3,4))
arr1 + arr2
```

```python
arr1 = np.random.randint(1,20, (3,4))
arr2 = np.random.randint(1,20, (3))
arr1 + arr2
```

```python
arr1 = np.random.randint(1,20, (1,3))
arr2 = np.random.randint(1,20, (4,3))
arr1 + arr2
```

```python
arr1 = np.random.randint(1,20, (1))
arr2 = np.random.randint(1,20, (2,2))
arr1 + arr2
```
