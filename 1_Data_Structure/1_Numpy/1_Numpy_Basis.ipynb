# hstack

```python
arr1 = np.random.random(6).reshape(2,3) * 25
arr2 = np.random.random(6).reshape(2,3) * 25
```

```python
hMergeArr = np.hstack((arr1, arr2))
hMergeArr
```

# hsplit

```python
np.hsplit(hMergeArr, 3)
```

# vstack

```python
vMergeArr = np.vstack((arr1, arr2))
vMergeArr
```

# vsplit

```python
np.vsplit(vMergeArr, 2)
```
