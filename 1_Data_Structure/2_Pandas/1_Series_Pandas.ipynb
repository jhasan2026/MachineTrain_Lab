# List to Series

```python
ser = pd.Series([1,2,3,4,5], name="Value form the array")
ser
```

```python
marks = [12,16,30,40]
subjects = ['Math', 'Physics', 'Bio', 'History']
df = pd.Series(marks, name='Jehan Marks',index=subjects)
df
```

# Series from Dictionary

```python
marksDictionary = {
    'maths':67,
    'english':57,
    'science':89,
    'bangla':100
}
df = pd.Series(marksDictionary, name='Jehan Marks')
df
```
